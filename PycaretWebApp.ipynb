```python
import pandas as pd
data = pd.read_csv("datasets/datasets_insurance.csv")
data.head()
```

```python
from pycaret.regression import *
exp = setup(data, target='charges', session_id = 123, 
             normalize = True, polynomial_features = True, trigonometry_features = True,
            feature_interaction=True, bin_numeric_features= ['age', 'bmi'])
```

```python
Linear_reg = create_model('lr')
```

```python
plot_model(Linear_reg, plot='residuals')
```
